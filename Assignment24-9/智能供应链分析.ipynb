```python
import matplotlib.pyplot as plt
import seaborn as sns
#相关性热力图
plt.figure(figsize=(40,20))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

```python
temp = pd.DatetimeIndex(df['order date (DateOrders)'])
# 获取不同尺度的时间
df['order_year'] = temp.year
df['order_month'] = temp.month
df['order_week_day'] = temp.weekday
df['order_hour'] = temp.hour
df['order_month_year'] = temp.to_period('M')
```

```python
df.head(10)
```

```python
plt.figure(figsize=(10,12))
plt.subplot(2,2,1)
df_year = df.groupby('order_year')
df_year['Sales'].mean().plot(figsize=(10,7),title='Average Sales in years')
plt.subplot(2,2,2)
df_week_day = df.groupby('order_week_day')
df_week_day['Sales'].mean().plot(figsize=(10,7),title='Average Sales in week day')
plt.subplot(2,2,3)
df_hour = df.groupby('order_hour')
df_hour['Sales'].mean().plot(figsize=(10,7),title='Average Sales in hours')
plt.subplot(2,2,4)
df_month = df.groupby('order_month')
df_month['Sales'].mean().plot(figsize=(10,7),title='Average Sales in months')
```

```python
#order region 分布
region = df.groupby('Order Region')
region['Sales per customer'].sum().sort_values(ascending=False).plot.bar(figsize=(12,7), title='Sum in different regions')
```

```python
#不同地区的销售额小提琴图
plt.figure(dpi=150)
sns.violinplot(x= "Order Region",y="Sales per customer",data=df,
               split=True,
               linewidth = 1, #线宽
               width = 1.6,   #箱之间的间隔比例
               palette = 'muted', #设置调色板
               order = df['Order Region'].unique().tolist(), #筛选类别
               # scale = 'count',  #测度小提琴图的宽度： area-面积相同,count-按照样本数量决定宽度,width-宽度一样
               gridsize = 50, #设置小提琴图的平滑度，越高越平滑
               # inner = 'box', #设置内部显示类型 --> 'box','quartile','point','stick',None
               #bw = 0.8      #控制拟合程度，一般可以不设置
               )
plt.show()
```
