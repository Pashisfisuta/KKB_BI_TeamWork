```python
import pandas as pd
data = pd.read_csv('data/data54641/voice.csv',encoding='utf-8')
data.head(10)
```

```python
#查看信息和缺失值
print(data.shape)
print(data.info())
data.describe()
```

```python
#看一下标签的分布
print('总样本量：{}'.format(data.shape[0]))
print('男性样本量：{}'.format(data[data.label=='male'].shape[0]))
print('女性样本量：{}'.format(data[data.label=='female'].shape[0]))

data.label.value_counts()
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
#使用标签编码
from sklearn.preprocessing import LabelEncoder,StandardScaler
gender_encoder = LabelEncoder()
y = gender_encoder.fit_transform(y)
#对x进行标准化
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
#划分训练和测试集，采用0.25作为测试集样本
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25)
```

```python
#加载训练所需的包
from sklearn.metrics import accuracy_score
from sklearn import metrics
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import mean_absolute_error,  make_scorer, accuracy_score
from catboost import CatBoostClassifier
from sklearn.model_selection import cross_val_score, learning_curve,train_test_split, GridSearchCV
```

```python
#xgboost
from xgboost import XGBClassifier
# 预设各参数的不同选项值
max_depth = [5,10,15]
learning_rate = [0.1,0.2,0.5]
n_estimators = [50,100,500]
min_child_weight = [2,3,5]
# 将各参数值以字典形式组织起来
parameters = {'max_depth':max_depth, 'learning_rate':learning_rate, 'n_estimators':n_estimators, 
              'min_child_weight':min_child_weight}
# 网格搜索法，测试不同的参数值
grid_dtcateg = GridSearchCV(estimator = XGBClassifier(booster= 'gbtree',random_state=1,objective='binary:logistic'),
                            param_grid = parameters, cv=10)
# 模型拟合
grid_dtcateg.fit(x_train, y_train)
# 返回最佳组合的参数值
print(grid_dtcateg.best_params_)

xgb = XGBClassifier(booster='gbtree',objective='binary:logistic',max_depth=5,learning_rate=0.05,n_estimators=300,min_child_weight=3,
                   random_state=7)
xgb_classifier = xgb.fit(x_train, y_train)
# 模型在测试集上的预测
y_pred = xgb_classifier.predict(x_test)
# 模型的准确率
print('XGBoost模型在测试集的预测准确率：%0.4lf\n' % metrics.accuracy_score(y_test, y_pred))
```

```python
#lightgbm
lgb_clf=LGBMClassifier(
                    n_jobs=-1,
                    n_estimators=500,
                    learning_rate=0.05,
                    max_depth=16,
                    num_leaves=10,
                    colsample_bytree=0.51,
                    subsample=0.6,
                    random_state=1
                )

scores = cross_val_score(lgb_clf, x_train, y_train, verbose=1, cv = 5, scoring=make_scorer(accuracy_score), n_jobs=-1)
scores.mean()
```

```python
#catboost
catboosting=CatBoostClassifier(iterations=10000, 
                              depth=7, 
                              learning_rate=0.02, 
                              loss_function='Logloss', 
                              eval_metric='AUC',
                              logging_level='Verbose', 
                              metric_period=500,
                              random_state=7)
catboosting.fit(x_train,y_train)
y_pred=catboosting.predict(x_test)
print('catboost模型在测试集的预测准确率：%0.4lf' % metrics.accuracy_score(y_test, y_pred))
```

```python
#SVM
from sklearn.metrics import accuracy_score
from sklearn import metrics
from sklearn.svm import LinearSVC
svm = LinearSVC()
svm_model = svm.fit(x_train, y_train)
# 模型在测试集上的预测
y_pred = svm_model.predict(x_test)
# 模型的准确率
print('SVM模型在测试集的预测准确率：%0.4lf' % metrics.accuracy_score(y_test, y_pred))
```
