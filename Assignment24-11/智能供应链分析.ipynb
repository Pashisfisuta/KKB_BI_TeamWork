```python
def R_Score(a,b,c):
    if a <= c[b][0.25]:
        return 4
    if a <= c[b][0.50]:
        return 3
    if a <= c[b][0.75]:
        return 2
    return 1
# Frequency Monetaryè¶Šå¤§è¶Šå¥½
def FM_Score(a,b,c):
    if a <= c[b][0.25]:
        return 1
    if a <= c[b][0.50]:
        return 2
    if a <= c[b][0.75]:
        return 3
    return 4
```

```python
from sklearn.metrics import accuracy_score,recall_score,roc_auc_score,confusion_matrix,f1_score
def model_stats(model,x_train,x_test,y_train,y_test,name='Fraud'):
    model = model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_pred,y_test)
    recall = recall_score(y_pred,y_test)
    auc = roc_auc_score(y_pred,y_test)
    f1 = f1_score(y_pred,y_test)
    confusion = confusion_matrix(y_pred,y_test) 
    print('Model Used:',model)
    print('{} Accuracy: {}%'.format(name,accuracy*100))
    print('{} Recall: {}%'.format(name,recall*100))
    print('{} AUC: {}%'.format(name,auc*100))
    print('{} F1 Score: {}%'.format(name,f1*100))
    print('{} Confusion Matrix:\n{}%'.format(name,confusion*100))
    return accuracy,recall,f1
```

```python
def rfm_user(df):
    """æ ¹æ®ç”¨æˆ·çš„R_soreã€F_scoreã€M_scoreçš„é«˜ä½Žè¿›è¡Œç”¨æˆ·å±‚çº§åˆ’åˆ†"""
    user_class = ''
    if df['R_score'] > 2 and df['F_score'] > 2 and df['M_score'] > 2:
        user_class = 'é‡è¦ä»·å€¼ç”¨æˆ·'
    elif df['R_score'] > 2 and df['F_score'] <= 2 and df['M_score'] > 2:
        user_class = 'é‡è¦å‘å±•ç”¨æˆ·'
    elif df['R_score'] <= 2 and df['F_score'] > 2 and df['M_score'] > 2:
        user_class = 'é‡è¦ä¿æŒç”¨æˆ·'
    elif df['R_score'] <= 2 and df['F_score'] <= 2 and df['M_score'] > 2:
        user_class = 'é‡è¦æŒ½ç•™ç”¨æˆ·'

    elif df['R_score'] > 2 and df['F_score'] > 2 and df['M_score'] <= 2:
        user_class = 'ä¸€èˆ¬ä»·å€¼ç”¨æˆ·'
    elif df['R_score'] > 2 and df['F_score'] <= 2 and df['M_score'] <= 2:
        user_class = 'ä¸€èˆ¬å‘å±•ç”¨æˆ·'
    elif df['R_score'] <= 2 and df['F_score'] > 2 and df['M_score'] <= 2:
        user_class = 'ä¸€èˆ¬ä¿æŒç”¨æˆ·'
    elif df['R_score'] <= 2 and df['F_score'] <= 2 and df['M_score'] <= 2:
        user_class = 'ä¸€èˆ¬æŒ½ç•™ç”¨æˆ·'
    return user_class

customer_strat['Customer_Segmentation'] = customer_strat.apply(rfm_user, axis=1)
customer_strat.head()
```

```python
def model_status(model, x_train, x_test, y_train, y_test):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print("æ¨¡åž‹é¢„æµ‹çš„å†³å®šç³»æ•°R^2:", model.score(x_test, y_test))
    mse = mean_squared_error(y_test, y_pred)
    print("æ¨¡åž‹é¢„æµ‹çš„MSE:", mse)
    print("æ¨¡åž‹é¢„æµ‹çš„RMSE:", mse**0.05)
    print("æ¨¡åž‹é¢„æµ‹çš„MAE:", mean_absolute_error(y_test, y_pred))
    return model
```
